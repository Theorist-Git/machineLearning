```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("../datasets/titanic.csv")
df.sample(5)
```

```python
# Usually, we go from left to right in univariate analysis, but here we'll start with numerical data
```

## 1. Categorical Data

### a. countplot

```python
sns.countplot(df, x = "Survived") # same as df["Survived"].value_counts().plot(kind='bar')
df["Survived"].value_counts()
```

```python
sns.countplot(df, x="Pclass")
df["Pclass"].value_counts()
```

### b. piechart

```python
df["Embarked"].value_counts().plot(kind="pie", autopct="%.2f")
```

## 2. Numerical Data

### a. histogram

```python
import matplotlib.pyplot as plt
plt.hist(df["Age"], bins=8)
```

### b. distplot

```python
# probability distribution function
sns.displot(df["Age"], kde=True, stat="density", kde_kws=dict(cut=3))
df["Age"].skew() # -ve implies left side skewed, 0 => normal, +ve => skewed right
```

### c. boxplot

```python
sns.boxplot(df["Age"], orient="y")
```

```python
print(df["Age"].min())
print(df["Age"].max())
print(df["Age"].mean())
```
