## 1. Scatterplot (Numerical v. Numerical) 

```python
tips["total_bill"].corr(tips["tip"])
sns.scatterplot(x=tips["total_bill"], y=tips["tip"], hue=tips["sex"], style=tips["smoker"], size=tips["size"])
```

## 2. Barplot (Categorical v. Numerical) 

```python
sns.barplot(x=titanic["Pclass"], y=titanic["Fare"], hue=titanic["Sex"])
titanic.groupby("Pclass")["Fare"].mean().round(2)
```

## 3. Box plot (categorical vs numerical)

```python
sns.boxplot(x=titanic["Sex"], y= titanic["Age"], hue=titanic["Survived"])
```

## 4. displot (categorical - numerical)

```python
sns.kdeplot(titanic[titanic["Survived"] == 0]["Age"])
sns.kdeplot(titanic[titanic["Survived"] == 1]["Age"])
```

## 5. Heat Map (Cat vs Cat)

```python
titanic.groupby(["Pclass", "Survived"])["Survived"].count()
pd.crosstab(titanic["Pclass"], titanic["Survived"])
sns.heatmap(pd.crosstab(titanic["Pclass"], titanic["Survived"]))
# df[df["Survived"]==1].count()
# print(f"Number survived in each class = ")
```
